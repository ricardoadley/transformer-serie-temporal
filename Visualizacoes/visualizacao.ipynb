```python
def remove_outlier(df):
  Q1 = np.percentile(df['total_w'].to_list(), 25)
  Q3 = np.percentile(df['total_w'].to_list(), 75)

  # Calcular os limites superior e inferior usando a fórmula fornecida
  limite_inferior = Q1 - 1.5 * (Q3 - Q1)
  limite_superior = Q3 + 1.5 * (Q3 - Q1)

  # Substituir outliers pelos limites
  df['total_w'] = [x if (x >= limite_inferior and x <= limite_superior) else limite_inferior if x < limite_inferior else limite_superior for x in df['total_w'].to_list()]
  return df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Suponha que df_lsd_filled seja seu DataFrame com datas na coluna 'Data' e 'total_w' sendo a série temporal

# Usar dados do DataFrame
dates = pd.to_datetime(df_lsd_filled['Data'])
r = df_lsd_filled['total_w'].values  # Valores da série temporal

# Mapear cada hora para o mês correspondente
months = dates.dt.month
hours = dates.dt.hour
angle_per_month = 2 * np.pi / 12
theta = [(angle_per_month * month) + ((angle_per_month / 24) * hour) for month, hour in zip(months, hours)]

# Plotar o gráfico de coordenadas polares
plt.figure(figsize=(8, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, r)

# Adicionar nome dos meses aos ângulos
month_names = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
ax.set_xticks(np.linspace(0, 2 * np.pi, 12, endpoint=False))
ax.set_xticklabels(month_names)

plt.title("Gráfico de Coordenadas Polares para Série Temporal por Hora com Meses")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dados
data = {
    'Modelo': ['Informer', 'Transformer Vanilla', 'XGBoost', 'Prophet', 'Arima'],
    'LSD': [9.54, 1.66, 0.94, 1.12, 1.17],
    'CEEI': [7.20, 1.47, 0.65, 1.13, 1.17],
    'CN': [7.07, 1.45, 0.54, 1.12, 1.18],
    'SPLAB': [7.20, 1.44, 0.64, 1.13, 1.16],
    'CTLABS': [7.20, 1.45, 0.61, 1.11, 1.17]
}

# Criando DataFrame
df = pd.DataFrame(data)

# Calculando o uso médio de memória de cada modelo
df_mean = df.set_index('Modelo').mean(axis=1).sort_values(ascending=False)

# Ordenando o DataFrame de acordo com o uso médio de memória
df_sorted = df_mean.index

# Transformando os dados em formato longo (para o seaborn)
df_long = df.melt(id_vars='Modelo', var_name='Base de Dados', value_name='Uso de Memória (GB)')
df_long['Modelo'] = pd.Categorical(df_long['Modelo'], categories=df_sorted, ordered=True)

sns.set(rc={'xtick.labelsize': 7, 'ytick.labelsize': 7}, palette='gray', style="whitegrid")

# Plotando os dados
plt.figure(figsize=(10, 4))
ax = sns.barplot(data=df_long, x='Modelo', y='Uso de Memória (GB)', hue='Base de Dados', palette='Greys')
plt.xlabel('')
plt.ylabel('Uso de Memória (GB)', fontsize=7)
plt.yticks(range(0, int(df_long['Uso de Memória (GB)'].max()) + 1))
plt.legend(title='Base de Dados', bbox_to_anchor=(1, 1), loc='upper left', fontsize=7, title_fontsize=7)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dados
data = {
    'Modelo': ['Informer', 'Transformer Vanilla', 'XGBoost', 'Prophet', 'Arima'],
    'LSD': [9.54, 1.66, 0.94, 1.12, 1.17],
    'CEEI': [7.20, 1.47, 0.65, 1.13, 1.17],
    'CN': [7.07, 1.45, 0.54, 1.12, 1.18],
    'SPLAB': [7.20, 1.44, 0.64, 1.13, 1.16],
    'CTLABS': [7.20, 1.45, 0.61, 1.11, 1.17]
}

# Criando DataFrame
df = pd.DataFrame(data)

# Transformando os dados em formato longo (para o seaborn)
df_long = df.melt(id_vars='Modelo', var_name='Base de Dados', value_name='Uso de Memória (GB)')

# Plotando os dados
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=df_long, x='Modelo', y='Uso de Memória (GB)', hue='Base de Dados', palette='Greys')
plt.xlabel('')
plt.ylabel('Uso de Memória (GB)',fontsize=7)
plt.yticks(range(0, int(df_long['Uso de Memória (GB)'].max()) + 1))
plt.legend(title='Base de Dados',bbox_to_anchor=(1, 1), loc='upper left')

# Adicionando os valores no topo de cada barra
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```
