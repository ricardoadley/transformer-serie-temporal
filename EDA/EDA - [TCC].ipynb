```python
def prepara_dataframe(df_path):
  df = pd.read_csv(df_path,sep=";")
  df['Data'] = pd.to_datetime(df['Data'])
  df.reset_index(drop=True,inplace=True)
  df['mes_ano'] = df['Data'].map(lambda x: 100*x.year + x.month)
  df['dia']  = df['Data'].dt.day
  df['mes']  = df['Data'].dt.month
  df['ano']  = df['Data'].dt.year
  return df
```

```python
def calc_registros_faltantes(df,qtd_registros_esperados):
  registros =  df.groupby(['mes_ano'])['Data'].count().reset_index(name='registros_no_mes')
  registros['registros_esperados'] = qtd_registros_esperados
  registros['registros_faltantes'] = registros['registros_esperados'] - registros['registros_no_mes']
  return registros
```

## Tamanho dos gaps

```python
def calc_gaps_size(df):
  df = df.sort_values(by='Data')
  df['Intervalo'] = df['Data'].diff()
  gap_threshold = pd.Timedelta(hours=1)
  gaps = df[df['Intervalo'] > gap_threshold]
  return gaps.shape[0], gaps
```

### LSD

Quanto tempo ficou sem registro ate esse registro acontecer

## preencher com foward fill o que faltou

```python
def filled_df(df):
  df = df[['Data','P - Total (W)']]
  df.rename(columns={'P - Total (W)':'total_w'},inplace=True)
  df['Data'] = pd.to_datetime(df['Data'])
  df = df.set_index('Data')
  df = df.resample('1H').ffill()
  df.reset_index(inplace=True)
  return df
```

```python
df_lsd_filled =filled_df(df_lsd)
df_ceei_filled = filled_df(df_ceei)
df_splab_filled = filled_df(df_splab)
df_cn_filled = filled_df(df_cn)
df_ctlabs_filled = filled_df(df_ctlabs)
```

usei o foward fill = https://datastud.dev/posts/time-series-resample

##Tratamento dos dados para novo csv

```python
def convert_values_to_float(df):
  df['total_w'] = df['total_w'].replace(',', '.', regex=True)
  df['total_w'] = df['total_w'].astype(float)
  return df
```

```python
df_lsd_filled = convert_values_to_float(df_lsd_filled)
df_ceei_filled = convert_values_to_float(df_ceei_filled)
df_spalb_filled = convert_values_to_float(df_splab_filled)
df_cn_filled = convert_values_to_float(df_cn_filled)
df_ctlabs_filled = convert_values_to_float(df_ctlabs_filled)
```
